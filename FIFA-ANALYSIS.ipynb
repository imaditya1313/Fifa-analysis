```python
data["Club"].fillna("No club",inplace=True)
data["Position"].fillna("unknown",inplace=True)
```

```python
def change(x):
    x=str(x)
    x = x.replace("€","")
    if "M" in x:
        x = float(x[:-1])*1000000
    elif "K" in x:
        x = float(x[:-1])*1000
    return float(x)

data["Wage"] = data["Wage"].apply(lambda y:change(y))
data["Value"] = data["Value"].apply(lambda y:change(y))
data["Release Clause"] = data["Release Clause"].apply(lambda y:change(y))

#As only Release Clause has null values we will fill null values with mean

data["Release Clause"].fillna(data["Release Clause"].mean(),inplace=True)
```

```python
data["International Reputation"].fillna(data["International Reputation"].mean(),inplace=True)
data["Skill Moves"].fillna(data["Skill Moves"].mean(),inplace=True)
data["Weak Foot"].fillna(data["Weak Foot"].mean(),inplace=True)
for i in data.iloc[:,53:87]:
    data[i].fillna(data[i].mean(),inplace=True)
```

```python
#here we remove lbs from weight fo it becomes numberic value and replace null values with mean.
def weight(x):
    if isinstance(x,str):
        return x[:-3]
    return x
data["Weight"] = data["Weight"].apply(weight).astype('float')
data["Weight"].fillna((data["Weight"].mean()),inplace=True)


#here we convert height to centimeters and replace null values with mean.
def height(x):
    s=None
    if isinstance(x,str):
        l = x.split("'")
        s = int(l[0])*30.48 + int(l[1])*2.54
    
    return s
data["Height"] = data["Height"].apply(height)
data["Height"].fillna((data["Height"].mean()),inplace=True)
```

```python
data['Joined'].fillna('0', inplace = True)
data['Jersey Number'].fillna("Not specified", inplace = True)
data['Contract Valid Until'].fillna("Not available", inplace = True)
```

```python
#we will fill all this values with the most common ie mode vaue of the data.(use data['Body Type'].value_counts())

data['Preferred Foot'].fillna((data['Preferred Foot'].mode()[0]),inplace=True)
data['Body Type'].fillna((data['Body Type'].mode()[0]),inplace=True)
data['Work Rate'].fillna((data['Work Rate'].mode()[0]),inplace=True)
```

```python
def skill(x):
    if type(x)==str:
        a = x.split("+")
        return int(a[0])+int(a[1])
    else:
        return x
    
arr = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
        'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 
        'LWB', 'LDM','CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
#LS - Left Striker, ST - Striker, RS - Right Striker, LW - Left Winger, LF - Left Forward, CF - Center Forward, RF - Right Forward, RW - Right Winger

for y in arr:
    data[y] = data[y].apply(lambda y:skill(y))
    data[y].fillna(0.0,inplace=True)
```

```python
# creating new features by aggregating the features

def defending(data):
    return int(round((data[['Marking', 'StandingTackle', 
                               'SlidingTackle']].mean()).mean()))

def general(data):
    return int(round((data[['HeadingAccuracy', 'Dribbling', 'Curve', 
                               'BallControl']].mean()).mean()))

def mental(data):
    return int(round((data[['Aggression', 'Interceptions', 'Positioning', 
                               'Vision','Composure']].mean()).mean()))

def passing(data):
    return int(round((data[['Crossing', 'ShortPassing',
                               'LongPassing']].mean()).mean()))

def mobility(data):
    return int(round((data[['Acceleration', 'SprintSpeed', 
                               'Agility','Reactions']].mean()).mean()))
def power(data):
    return int(round((data[['Balance', 'Jumping', 'Stamina', 
                               'Strength']].mean()).mean()))

def rating(data):
    return int(round((data[['Potential', 'Overall']].mean()).mean()))

def shooting(data):
    return int(round((data[['Finishing', 'Volleys', 'FKAccuracy', 
                               'ShotPower','LongShots', 'Penalties']].mean()).mean()))
```

```python
# adding these categories to the data

data['Defending'] = data.apply(defending, axis = 1)
data['General'] = data.apply(general, axis = 1)
data['Mental'] = data.apply(mental, axis = 1)
data['Passing'] = data.apply(passing, axis = 1)
data['Mobility'] = data.apply(mobility, axis = 1)
data['Power'] = data.apply(power, axis = 1)
data['Rating'] = data.apply(rating, axis = 1)
data['Shooting'] = data.apply(shooting, axis = 1)

# lets check the column names in the data after adding new features
data.columns
```

```python
from sklearn.model_selection import train_test_split 
from sklearn import metrics
```

```python
X = data[['Age','Potential','GKHandling','Penalties','Composure','Marking','StandingTackle','SlidingTackle','GKPositioning','GKReflexes']]
y = data['Overall']
y.shape
```

```python
#accuracy of testing prediction i.e R-squared
acc_LR=lin_regressor.score(X_test,y_test)
print(acc_LR)
```
